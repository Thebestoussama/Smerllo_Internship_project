```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Transactions.csv')
```

```python
df = df.sample(100000 , random_state = 42)
```

```python
df['instance_date'] = pd.to_datetime(df['instance_date'])
```

```python
def flag_property(row):
    if row["property_type_en"] == "Unit" and row["property_usage_en"] == "Residential":
        return "Apartment"
    elif row["property_type_en"] == "Villa" and row["property_usage_en"] == "Residential":
        return "Villa"
    elif row['property_type_en'] == 'Unit'and row['property_usage_en'] == 'Commercial':
        return "Commercial"
    elif row['property_type_en'] == 'Villa'and row['property_usage_en'] == 'Commercial':
        return "Commercial"
    elif row["property_type_en"] == "Land":
        return "Plot"
    else:
        return "Other"


df["category_type"] = df.apply(flag_property, axis=1)
```

```python
df['year'] = df['instance_date'].dt.year
df['month'] = df['instance_date'].dt.month
```

```python
df['actual_worth'] = df['actual_worth'].fillna(np.nan)
```

```python
filter_df = df.groupby(['category_type'])
    
df1 = filter_df.get_group('Apartment')
mean1 = df1['actual_worth'].mean()
    
df2 = filter_df.get_group('Villa')
mean2 = df2['actual_worth'].mean()
    
df3 = filter_df.get_group('Commercial')
mean3 = df3['actual_worth'].mean()
    
df4 = filter_df.get_group('Other')
mean4 = df4['actual_worth'].mean()
    
df5 = filter_df.get_group('Plot')
mean5 = df5['actual_worth'].mean()
    
```

```python
df['actual_worth'] = df['actual_worth'].fillna(0)
```

```python
def flag_actual_worth(row):
    if row['category_type'] == 'Apartment': 
        if row['actual_worth'] == 0  :
            return mean1
        else:
            return row['actual_worth']
        
    if row['category_type'] == 'Villa': 
        if row['actual_worth'] == 0 :
            return mean2 
        else:
            return row['actual_worth']
    
    if row['category_type'] == 'Commercial' :
        if row['actual_worth'] == 0 :
            return mean3      
        else:
            return row['actual_worth']
    
    if row['category_type'] == 'Other' :
        if row['actual_worth'] == 0 :
            return mean4 
        else:
            return row['actual_worth']
    
    if row['category_type'] == 'Plot' :
        if row['actual_worth'] == 0 :
            return mean5  
        else:
            return row['actual_worth']
    

df['transaction_value'] = df.apply(flag_actual_worth , axis = 1)
```

```python
def flag_reg_type(row):
    
    if row['reg_type_en'] == 'Off-Plan Properties' :
        
        return 'Offplan'
    
    else :
        
        return 'Ready'
    
df['structure'] = df.apply(flag_reg_type , axis = 1)   
  
```

```python
agg1_df = df.groupby(['month','category_type' , 'area_name_en', 'trans_group_en', 'structure'])

          
YoY_dict = {'YoY%' : [0.]}

def KPIs(month , category_type , area, procedure, structure) :
    
    df_plot = {'year' : [] ,
          'Yearly sales volume in all areas' : [],
          'Yearly sales value in all areas' : []
          }
    
    YoY_dict = {'YoY%' : [0.]}
    
    df1 = agg1_df.get_group((month , category_type , area, procedure, structure))
    agg2_df = df1.groupby(['year'])
    
    for year in range(1997, 2024):
        
        try : 
            df2 = agg2_df.get_group(year)
            df_plot['Yearly sales volume in all areas'].append(len(df2.index))
            df_plot['Yearly sales value in all areas'].append(df2['transaction_value'].sum())
            df_plot['year'].append(year)
            
        except KeyError:
            df_plot['Yearly sales volume in all areas'].append(0)
            df_plot['Yearly sales value in all areas'].append(0)
            df_plot['year'].append(year)
            
    
    df_plot = pd.DataFrame(df_plot)
    df_plot = df_plot.set_index('year')
    l_index = df_plot.index
    
    for year in l_index[1:] :
        
        Num_transactions_year_last = df_plot['Yearly sales volume in all areas'][year-1]
        Num_transactions_year_current = df_plot['Yearly sales volume in all areas'][year]
        
        if Num_transactions_year_last != 0 :
            YoY = ((Num_transactions_year_current - Num_transactions_year_last)/ (Num_transactions_year_last))*100
            YoY_dict['YoY%'] .append(YoY)
        else :
            YoY_dict['YoY%'].append(0) # 0 mean that there is no transaction in the last year
            
    YoY_dict = pd.DataFrame(YoY_dict)  
    YoY_dict = YoY_dict.set_index(df_plot.index)
    
    df_plot = pd.concat([df_plot[['Yearly sales volume in all areas','Yearly sales value in all areas']] , YoY_dict] , axis = 1)
    
    
    return df_plot
        
   
```

```python
#KPIs(7 , 'Villa' , 'Hadaeq Sheikh Mohammed Bin Rashid' , 'Sales' , 'Ready')
```

```python
def plot_KPIs(month , category_type , area, procedure, structure) :
    
    data_plot = KPIs(month , category_type , area, procedure, structure)
    
    #fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(30, 20))
    
    fig, ax1 = plt.subplots(figsize=(20, 10))  
    ax2 = plt.subplots(figsize=(20, 10))[1]   
    ax3 = plt.subplots(figsize=(20, 10))[1]
    
    sns.barplot(data = data_plot , x = data_plot.index , y = 'Yearly sales volume in all areas' , ax = ax1 , color ='g' )
    sns.barplot(data = data_plot , x = data_plot.index , y = 'Yearly sales value in all areas' , ax = ax2 , color ='b' )
    ax3.plot(data_plot.index , data_plot['YoY%'] , color = 'r')
    
    
    ax1.set_xlabel('Years')
    ax1.set_ylabel('Number of transactions')
    ax1.set_title('Yearly sales volume in all areas')
    ax1.grid(True, linestyle='dotted')

    
    ax2.set_xlabel('Years')
    ax2.set_ylabel('Yearly sales value in all areas')
    ax2.set_title('Yearly sales value in all areas')
    ax2.grid(True, linestyle='dotted')
    
    
    ax3.set_xticks(data_plot.index[::1], ha='right')
    ax3.set_xlabel('Years')
    ax3.set_ylabel('Year_Over_Year')
    ax3.grid(True, linestyle='dotted')
    ax3.set_title('Property price change in all areas')
    
    plt.tight_layout()
    
    plt.show() 
    
    
    
```
