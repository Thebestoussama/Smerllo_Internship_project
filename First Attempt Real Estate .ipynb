```python
FEATURES = ['trans_group_en','property_type_en', 'property_usage_en', 'has_parking', 'procedure_area' ,'actual_worth', 'meter_sale_price']
df = df[FEATURES]
df
```

```python
df['trans_group_en'] = df['trans_group_en'].astype('category').cat.codes 
df['property_type_en'] = df['trans_group_en'].astype('category').cat.codes 
df['property_usage_en'] = df['trans_group_en'].astype('category').cat.codes 
```

```python
X_train = df.drop('meter_sale_price', axis=1)
y_train = df['meter_sale_price']
```

```python
X_train
```

```python
y_train  = pd.DataFrame(y_train)
y_train
```

```python
reg =xgb.XGBRegressor(base_score=0.5,
                       booster='gbtree',    
                       n_estimators=1500,
                       objective='reg:linear',
                       max_depth=10,
                       learning_rate=0.01)
reg.fit(X_train, y_train,
        eval_set=[(X_train, y_train)],
        verbose=1)
```
