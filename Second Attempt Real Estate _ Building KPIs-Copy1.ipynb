```python
FEATURES = ['transaction_year', 'transaction_month', 'transaction_category',
       'property_type', 'property_usage', 'property_off_plan', 'district',
       'building_name', 'property_area', 'sales_value', 'meter_sale_price',
       'rent_value', 'meter_rent_price']
data = df[FEATURES]
```

```python
for i in range(0,1000):
    if df.loc[i,'property_type'] == 'Unit'and df.loc[i,'property_usage'] == 'Residential':
        df['category_type'] = df['category_type'].replace({df.loc[i,'category_type'] : 'Apartement'})
        
    elif df.loc[i,'property_type'] == 'Villa'and df.loc[i,'property_usage'] == 'Residential':
        df['category_type'] = df['category_type'].replace({df.loc[i,'category_type'] : 'Villa'})
        
    elif df.loc[i,'property_type'] == 'Villa'and df.loc[i,'property_usage'] == 'Commercial':
        df['category_type'] = df['category_type'].replace({df.loc[i,'category_type'] : 'Commercial'})
        
    elif df.loc[i,'property_type'] == 'Unit'and df.loc[i,'property_usage'] == 'Commercial':
        df['category_type'] = df['category_type'].replace({df.loc[i,'category_type'] : 'Commercial'})
        
    elif df.loc[i,'property_type'] == 'Land':
        df['category_type'] = df['category_type'].replace({df.loc[i,'category_type'] : 'Plot'})
        
    elif df.loc[i,'property_type'] == 'Building':
        df['category_type'] = df['category_type'].replace({df.loc[i,'category_type'] : 'Building'})
        
        

df = pd.concat([data, pd.DataFrame(df['category_type'])], axis=1)  
```

```python
mon_dict = {}
grouped_df = df.groupby(['transaction_year','category_type'])
for k in range(1999,2024):
    
    try: 
        mon_dict[f'transaction_year_Apartement_{k}'] = grouped_df.get_group((k, 'Apartement'))
        mon_dict[f'transaction_year_Villa_{k}'] = grouped_df.get_group((k, 'Villa'))
        mon_dict[f'transaction_year_Commercial_{k}'] = grouped_df.get_group((k, 'Commercial'))
        mon_dict[f'transaction_year_Building_{k}'] = grouped_df.get_group((k, 'Building'))
        mon_dict[f'transaction_year_Plot_{k}'] = grouped_df.get_group((k, 'Plot'))
    except KeyError: 
        try:
            mon_dict[f'transaction_year_Villa_{k}'] = grouped_df.get_group((k, 'Villa'))
            mon_dict[f'transaction_year_Commercial_{k}'] = grouped_df.get_group((k, 'Commercial'))
            mon_dict[f'transaction_year_Building_{k}'] = grouped_df.get_group((k, 'Building'))
            mon_dict[f'transaction_year_Plot_{k}'] = grouped_df.get_group((k, 'Plot'))
        except KeyError:
            try:
                mon_dict[f'transaction_year_Commercial_{k}'] = grouped_df.get_group((k, 'Commercial'))
                mon_dict[f'transaction_year_Building_{k}'] = grouped_df.get_group((k, 'Building'))
                mon_dict[f'transaction_year_Plot_{k}'] = grouped_df.get_group((k, 'Plot'))
            except KeyError:
                try:
                    mon_dict[f'transaction_year_Building_{k}'] = grouped_df.get_group((k, 'Building'))
                    mon_dict[f'transaction_year_Plot_{k}'] = grouped_df.get_group((k, 'Plot'))
                except KeyError:
                    try: 
                        mon_dict[f'transaction_year_Plot_{k}'] = grouped_df.get_group((k, 'Plot'))
                    except KeyError:
                        continue
        
       
```

```python
dict_init_Villa = {'Number of Transactions': 0 ,
             'Change in Number of Transactions': 0,
             'Sales Value': 0,
             'Change in Sales Value':0 ,
             'Average Price': 0,
             'Change in Average Price': 0
            }
dict_init_Apartement = {'Number of Transactions': 0 ,
             'Change in Number of Transactions': 0,
             'Sales Value': 0,
             'Change in Sales Value':0 ,
             'Average Price': 0,
             'Change in Average Price': 0
            }
dict_init_Commercial = {'Number of Transactions': 0 ,
             'Change in Number of Transactions': 0,
             'Sales Value': 0,
             'Change in Sales Value':0 ,
             'Average Price': 0,
             'Change in Average Price': 0
            }
dict_init_Plot = {'Number of Transactions': 0 ,
             'Change in Number of Transactions': 0,
             'Sales Value': 0,
             'Change in Sales Value':0 ,
             'Average Price': 0,
             'Change in Average Price': 0
            }
dict_init_Building = {'Number of Transactions': 0 ,
             'Change in Number of Transactions': 0,
             'Sales Value': 0,
             'Change in Sales Value':0 ,
             'Average Price': 0,
             'Change in Average Price': 0
            }

kpi_dict = {'Villa': dict_init_Villa ,
            'Apartement' : dict_init_Apartement ,
            'Commercial' : dict_init_Commercial ,
            'Plot' : dict_init_Plot ,
            'Building' : dict_init_Building
           }
```
