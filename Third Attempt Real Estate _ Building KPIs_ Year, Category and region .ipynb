```python
FEATURES = ['transaction_id', 'procedure_id', 'trans_group_id', 'trans_group_ar',
       'trans_group_en', 'procedure_name_ar', 'procedure_name_en',
       'instance_date', 'property_type_id', 'property_type_ar',
       'property_type_en', 'property_sub_type_id', 'property_sub_type_ar',
       'property_sub_type_en', 'property_usage_ar', 'property_usage_en',
       'reg_type_id', 'reg_type_ar', 'reg_type_en', 'area_id', 'area_name_ar',
       'area_name_en', 'building_name_ar', 'building_name_en',
       'project_number', 'project_name_ar', 'project_name_en',
       'master_project_en', 'master_project_ar', 'nearest_landmark_ar',
       'nearest_landmark_en', 'nearest_metro_ar', 'nearest_metro_en',
       'nearest_mall_ar', 'nearest_mall_en', 'rooms_ar', 'rooms_en',
       'has_parking', 'procedure_area', 'actual_worth', 'meter_sale_price',
       'rent_value', 'meter_rent_price', 'no_of_parties_role_1',
       'no_of_parties_role_2', 'no_of_parties_role_3', 'year']

data = df[FEATURES]
```

```python
for i in range(0,1000):
    if df.loc[i,'property_type_en'] == 'Unit'and df.loc[i,'property_usage_en'] == 'Residential':
        df['category_type'] = df['category_type'].replace({df.loc[i,'category_type'] : 'Apartement'})
        
    elif df.loc[i,'property_type_en'] == 'Villa'and df.loc[i,'property_usage_en'] == 'Residential':
        df['category_type'] = df['category_type'].replace({df.loc[i,'category_type'] : 'Villa'})
        
    elif df.loc[i,'property_type_en'] == 'Villa'and df.loc[i,'property_usage_en'] == 'Commercial':
        df['category_type'] = df['category_type'].replace({df.loc[i,'category_type'] : 'Commercial'})
        
    elif df.loc[i,'property_type_en'] == 'Unit'and df.loc[i,'property_usage_en'] == 'Commercial':
        df['category_type'] = df['category_type'].replace({df.loc[i,'category_type'] : 'Commercial'})
        
    elif df.loc[i,'property_type_en'] == 'Land':
        df['category_type'] = df['category_type'].replace({df.loc[i,'category_type'] : 'Plot'})
    else:
        df['category_type'] = df['category_type'].replace({df.loc[i,'category_type'] : 'Other'})
        
        

df = pd.concat([data, pd.DataFrame(df['category_type'])], axis=1) 
```

# Groupby year, category_type and area in a dictionary

```python
df['area_name_en'].unique()
```

```python
len(df['area_name_en'].unique())
```

```python
mon_dict = {}
grouped_df = df.groupby(['year','category_type' , 'area_name_en'])
for k in range(1997,2024):
    mon_dict[f'transaction_Apartement_{k}'], mon_dict[f'transaction_Villa_{k}'], mon_dict[f'transaction_Commercial_{k}'],mon_dict[f'transaction_Other_{k}'], mon_dict[f'transaction_Plot_{k}'] = {},{},{},{},{} 
    for i in range(0,len(df['area_name_en'].unique())):
        try:
            mon_dict[f'transaction_Apartement_{k}'][df['area_name_en'].unique()[i]] = grouped_df.get_group((k, 'Apartement', df['area_name_en'].unique()[i]))
            mon_dict[f'transaction_Villa_{k}'][df['area_name_en'].unique()[i]] = grouped_df.get_group((k, 'Villa', df['area_name_en'].unique()[i]))
            mon_dict[f'transaction_Commercial_{k}'][df['area_name_en'].unique()[i]] = grouped_df.get_group((k, 'Commercial', df['area_name_en'].unique()[i]))
            mon_dict[f'transaction_Other_{k}'][df['area_name_en'].unique()[i]] = grouped_df.get_group((k, 'Other', df['area_name_en'].unique()[i]))
            mon_dict[f'transaction_Plot_{k}'][df['area_name_en'].unique()[i]] = grouped_df.get_group((k, 'Plot', df['area_name_en'].unique()[i]))
        except KeyError: 
            try:
                mon_dict[f'transaction_Villa_{k}'][df['area_name_en'].unique()[i]] = grouped_df.get_group((k, 'Villa', df['area_name_en'].unique()[i]))
                mon_dict[f'transaction_Commercial_{k}'][df['area_name_en'].unique()[i]] = grouped_df.get_group((k, 'Commercial', df['area_name_en'].unique()[i]))
                mon_dict[f'transaction_Other_{k}'][df['area_name_en'].unique()[i]] = grouped_df.get_group((k, 'Other', df['area_name_en'].unique()[i]))
                mon_dict[f'transaction_Plot_{k}'][df['area_name_en'].unique()[i]] = grouped_df.get_group((k, 'Plot', df['area_name_en'].unique()[i]))
            except KeyError:
                try:
                    mon_dict[f'transaction_Commercial_{k}'][df['area_name_en'].unique()[i]] = grouped_df.get_group((k, 'Commercial', df['area_name_en'].unique()[i]))
                    mon_dict[f'transaction_Other_{k}'][df['area_name_en'].unique()[i]] = grouped_df.get_group((k, 'Other', df['area_name_en'].unique()[i]))
                    mon_dict[f'transaction_Plot_{k}'][df['area_name_en'].unique()[i]] = grouped_df.get_group((k, 'Plot', df['area_name_en'].unique()[i]))
                except KeyError:
                    try:
                        mon_dict[f'transaction_Other_{k}'][df['area_name_en'].unique()[i]] = grouped_df.get_group((k, 'Other', df['area_name_en'].unique()[i]))
                        mon_dict[f'transaction_Plot_{k}'][df['area_name_en'].unique()[i]] = grouped_df.get_group((k, 'Plot', df['area_name_en'].unique()[i]))
                    except KeyError:
                        try:
                            mon_dict[f'transaction_Plot_{k}'][df['area_name_en'].unique()[i]] = grouped_df.get_group((k, 'Plot', df['area_name_en'].unique()[i]))
                        except KeyError:
                            continue

    
   
```

```python
def KPIs(info , region ) : 
    kpi = {}
    if mon_dict[info] != {} :
        kpi = {'Num_transactions' : len(mon_dict[info][region].index),
                                       'Sales_Value' : mon_dict[info][region]['meter_sale_price'].sum()
                                      }  
        return kpi
    else : 
        print('there is no transaction that matches your details ')
        
```
